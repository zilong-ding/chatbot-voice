```python
import sys
from PyQt5.QtCore import *
from PyQt5.QtGui import *
from PyQt5.QtWidgets import *

def window():
   app = QApplication(sys.argv)
   w = QWidget()
   b = QPushButton(w)
   b.setText("Show message!")

   b.move(100,50)
   b.clicked.connect(showdialog)
   w.setWindowTitle("PyQt MessageBox demo")
   w.show()
   sys.exit(app.exec_())

def showdialog():
   msg = QMessageBox()
   msg.setIcon(QMessageBox.Information)

   msg.setText("This is a message box")
   msg.setInformativeText("This is additional information")
   msg.setWindowTitle("MessageBox demo")
   msg.setDetailedText("The details are as follows:")
   msg.setStandardButtons(QMessageBox.Ok | QMessageBox.Cancel)
   msg.buttonClicked.connect(msgbtn)

   retval = msg.exec_()

def msgbtn(i):
   print ("Button pressed is:",i.text())

if __name__ == '__main__':
   window()
```

```python
import sys
from PyQt5.QtCore import *
from PyQt5.QtGui import *
from PyQt5.QtWidgets import *

class MainWindow(QMainWindow):
   count = 0

   def __init__(self, parent = None):
      super(MainWindow, self).__init__(parent)
      self.mdi = QMdiArea()
      self.setCentralWidget(self.mdi)
      bar = self.menuBar()

      file = bar.addMenu("File")
      file.addAction("New")
      file.addAction("cascade")
      file.addAction("Tiled")
      file.triggered[QAction].connect(self.windowaction)
      self.setWindowTitle("MDI demo")

   def windowaction(self, q):
      print ("triggered")

      if q.text() == "New":
         MainWindow.count = MainWindow.count+1
         sub = QMdiSubWindow()
         sub.setWidget(QTextEdit())
         sub.setWindowTitle("subwindow"+str(MainWindow.count))
         self.mdi.addSubWindow(sub)
         sub.show()

      if q.text() == "cascade":
         self.mdi.cascadeSubWindows()

      if q.text() == "Tiled":
         self.mdi.tileSubWindows()

def main():
   app = QApplication(sys.argv)
   ex = MainWindow()
   ex.show()
   sys.exit(app.exec_())

if __name__ == '__main__':
   main()
```

```python
import pyaudio
import wave
pa = pyaudio.PyAudio()
```

```python
# 打开声卡，设置 采样深度为16位、声道数为2、采样率为16、输入、采样点缓存数量为2048
stream = pa.open(format=pyaudio.paInt16, channels=2, rate=16000, input=True, frames_per_buffer=2048)
```

```python
# 新建一个列表，用来存储采样到的数据
record_buf = []
count = 0
while count < 8*5:
    audio_data = stream.read(2048)      # 读出声卡缓冲区的音频数据
    record_buf.append(audio_data)       # 将读出的音频数据追加到record_buf列表
    count += 1
    # print(audio_data)
print(record_buf)
```

```python
wf = wave.open('01.wav', 'wb')          # 创建一个音频文件，名字为“01.wav"

wf.setnchannels(2)                      # 设置声道数为2
wf.setsampwidth(2)                      # 设置采样深度为
wf.setframerate(16000)                  # 设置采样率为16000
 # 将数据写入创建的音频文件
wf.writeframes("".encode().join(record_buf))
print('ok')
 # 写完后将文件关闭
wf.close()
```

```python
import sys
from PyQt5.QtWidgets import QApplication, QWidget, QPushButton
from PyQt5.QtCore import QIODevice, QFile, QBuffer, QByteArray
from PyQt5.QtMultimedia import QAudioFormat, QAudioInput, QAudioOutput

class Example(QWidget):

    def __init__(self):
        super().__init__()

        self.initUI()

    def initUI(self):

        # 创建一个按钮
        self.btn = QPushButton('Hold me', self)
        self.btn.resize(self.btn.sizeHint())
        self.btn.move(50, 50)

        # 创建一个QAudioInput对象
        self.audio_input = QAudioInput()

        # 创建一个QAudioOutput对象
        self.audio_output = QAudioOutput()

        # 创建一个QBuffer对象，用于存储录制的音频数据
        self.buffer = QBuffer()
        self.buffer.open(QIODevice.ReadWrite)

        # 将按钮按下和释放事件连接到函数
        self.btn.pressed.connect(self.start_recording)
        self.btn.released.connect(self.stop_recording)

        self.setGeometry(300, 300, 250, 150)
        self.setWindowTitle('Example')
        self.show()

    def start_recording(self):
        # 设置音频输入格式
        format = self.audio_input.format()
        self.audio_input.setFormat(format)

        # 开始录制音频
        self.audio_input.start(self.buffer)

    def stop_recording(self):
        # 停止录制音频
        self.audio_input.stop()

        # 重置QBuffer的读写位置，以便播放录制的音频数据
        self.buffer.seek(0)

        # 设置音频输出格式
        format = self.audio_output.format()
        self.audio_output.setFormat(format)

        # 创建一个QByteArray对象，用于存储从QBuffer中读取的音频数据
        data = QByteArray()
        data.append(self.buffer.readAll())

        # 开始播放录制的音频
        self.audio_output.start()
        self.audio_output.write(data)

if __name__ == '__main__':
    app = QApplication(sys.argv)
    ex = Example()
    sys.exit(app.exec_())
```

```python
import sys
import pyaudio
import wave
from PyQt5.QtWidgets import QApplication, QWidget, QVBoxLayout, QPushButton

class AudioRecorder:
   def __init__(self):
       self.chunk = 1024
       self.format = pyaudio.paInt16
       self.channels = 1
       self.rate = 44100
       self.frames = []

       self.audio = pyaudio.PyAudio()

   def start_recording(self):
       self.stream = self.audio.open(format=self.format,
                                     channels=self.channels,
                                     rate=self.rate,
                                     input=True,
                                     frames_per_buffer=self.chunk)

   def stop_recording(self):
       self.stream.stop_stream()
       self.stream.close()

   def play_recording(self):
       playback_stream = self.audio.open(format=self.format,
                                         channels=self.channels,
                                         rate=self.rate,
                                         output=True,
                                         frames_per_buffer=self.chunk)
       for frame in self.frames:
           playback_stream.write(frame)
       playback_stream.stop_stream()
       playback_stream.close()

   def record(self, in_data, frame_count, time_info, status_flags):
       self.frames.append(in_data)
       return (in_data, pyaudio.paContinue)

class App(QWidget):
   def __init__(self):
       super().__init__()
       self.init_ui()

   def init_ui(self):
       self.recorder = AudioRecorder()

       layout = QVBoxLayout()

       self.record_button = QPushButton('按住录音，松开播放', self)
       self.record_button.pressed.connect(self.recorder.start_recording)
       self.record_button.released.connect(self.on_release)

       layout.addWidget(self.record_button)

       self.setLayout(layout)
       self.setWindowTitle('录音示例')
       self.show()

   def on_release(self):
       self.recorder.stop_recording()
       self.recorder.play_recording()
       self.recorder.frames = []

if __name__ == '__main__':
   app = QApplication(sys.argv)
   ex = App()
   sys.exit(app.exec_())
```

```python
import sys
from PyQt5.QtWidgets import QApplication, QWidget, QVBoxLayout, QPushButton
from PyQt5.QtCore import QIODevice, QByteArray
from PyQt5.QtMultimedia import QAudioInput, QAudioOutput, QAudioFormat, QAudioDeviceInfo

class AudioBuffer(QIODevice):
    def __init__(self, parent=None):
        super().__init__(parent)
        self.buffer = QByteArray()

    def writeData(self, data):
        self.buffer.append(data)
        return len(data)

    def readData(self, maxlen):
        data = self.buffer[:maxlen]
        self.buffer = self.buffer[maxlen:]
        return data

class App(QWidget):
    def __init__(self):
        super().__init__()
        self.init_ui()

    def init_ui(self):
        layout = QVBoxLayout()

        self.record_button = QPushButton('按住录音，松开播放', self)
        self.record_button.pressed.connect(self.start_recording)
        self.record_button.released.connect(self.stop_recording_and_play)

        layout.addWidget(self.record_button)

        self.setLayout(layout)
        self.setWindowTitle('录音示例')
        self.show()

        self.audio_format = QAudioFormat()
        self.audio_format.setSampleRate(44100)
        self.audio_format.setChannelCount(1)
        self.audio_format.setSampleSize(16)
        self.audio_format.setCodec("audio/pcm")
        self.audio_format.setByteOrder(QAudioFormat.LittleEndian)
        self.audio_format.setSampleType(QAudioFormat.SignedInt)

        self.audio_input = None
        self.audio_output = None
        self.audio_buffer = AudioBuffer()

    def start_recording(self):
        self.audio_buffer.buffer.clear()
        input_device_info = QAudioDeviceInfo.defaultInputDevice()
        self.audio_input = QAudioInput(input_device_info, self.audio_format, self)
        self.audio_input.start(self.audio_buffer)

    def stop_recording_and_play(self):
        self.audio_input.stop()
        self.audio_input.deleteLater()
        self.audio_input = None

        output_device_info = QAudioDeviceInfo.defaultOutputDevice()
        self.audio_output = QAudioOutput(output_device_info, self.audio_format, self)
        self.audio_output.start(self.audio_buffer)

if __name__ == '__main__':
    app = QApplication(sys.argv)
    ex = App()
    sys.exit(app.exec_())
```
